```python
def plot2Lists(l1,l2,title = "",ylabel = "", xlabel = ""):
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.plot(l1,l2)
    plt.show()
```

```python
def getMultiLabelEntityDistribution(dados, verbose = False):
    distributionDict = dict()
    for sent in dados:
        for tokano in sent:
            if tokano[1][0] != "O": #Retira o 'O'utside
                size = len(tokano[1])
                try:
                    distributionDict[size] += 1 
                except:
                    distributionDict[size] = 1 

    distribution = [(k,v) for k, v in distributionDict.items()]
    distribution.sort(key=lambda x: x[0])
    #print(distribution)
    
    word_rank,word_frequency = zip(*distribution)
    #print(word_frequency)
    
    if verbose == True:
        plot2Lists(word_rank,word_frequency,title = "Quantidade de Entidades Multilabel",ylabel = "Total Number of Occurrences", xlabel = "Entity Quantity by word")

        import matplotlib.pyplot as plt
        fig, ax = plt.subplots(figsize=(7,5), dpi=100)
        plt.barh(word_rank, width=word_frequency)
        
        for i, v in enumerate(word_frequency):
            #ax.text(v, i+0.9, str(v), color='black', fontweight='bold')
            ax.text(v, i+1, str(v), color='black', fontweight='bold')
    return distribution

```

```python
def getMultiLabelEntityDistribution(dados, verbose = False):
    distributionDict = dict()
    for sent in dados:
        for tokano in sent:
            if tokano[1][0] != "O": #Retira o 'O'utside
                size = len(tokano[1])
                try:
                    distributionDict[size] += 1 
                except:
                    distributionDict[size] = 1 

    distribution = [(k,v) for k, v in distributionDict.items()]
    distribution.sort(key=lambda x: x[0])
    
    
    word_rank,word_frequency = zip(*distribution)
    
    if verbose == True:
        plot2Lists(word_rank,word_frequency,title = "Quantidade de Entidades Multilabel",ylabel = "Total Number of Occurrences", xlabel = "Entity Quantity by word")

        import matplotlib.pyplot as plt
        fig, ax = plt.subplots(figsize=(7,5), dpi=100)
        plt.barh(word_rank, width=word_frequency)
        
        for i, v in enumerate(word_frequency):
            #ax.text(v, i+0.9, str(v), color='black', fontweight='bold')
            ax.text(v, i+0.9, str(v), color='black', fontweight='bold')

    return distribution
```

```python
import pandas as pd

def getMultiLabelSizeEntityMatrixDivided(dados,tam = 6, verbose = False):
    entityList = AC.getEntitiesUsed(dados)
    tamanho = [x for x in range(1,tam+1)]
    
    distributionDict = dict()
    
    for entityName in entityList:
        distributionDict[entityName] = [0 for x in range(len(tamanho))]
    
    for sent in dados:
        for tokano in sent:
            if tokano[1][0] != "O": #Retira o 'O'utside
                size = len(tokano[1]) - 1        
                for entityTag in tokano[1]:
                    entityName = entityTag[2:]
                    try:
                        distributionDict[entityName][size] += 1
                    except:
                        pass
                        #print(tokano[1])
                    
    distributionDictDivided = dict()
    for d in distributionDict.items():
        soma = sum(d[1])
        distributionDictDivided[d[0]] = [(st/soma)*100 for st in d[1]]
            
    distributionDict = distributionDictDivided
    #print(distributionDict)      
    #for p in distributionDict.items():
    #    print(p)
    
    #d = {'col1': [1, 2], 'col2': [3, 4]}
    df = pd.DataFrame(data=distributionDict)
    df = df.T #transpose
    df = df.round(1)
    return df,distributionDict
```
